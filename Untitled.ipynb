```python
import numpy as np
from scipy.signal import wiener

img = cv2.imread("uniform.jpg")
resized = cv2.resize(img, (0, 0), fx=0.2, fy=0.2)
resized = resized/255

# Converti l'immagine in un array numpy
image_array = np.array(resized)

# Applica il filtro di Wiener
filtered_image = wiener(image_array, (5,5))

# Ora puoi fare qualcosa con l'immagine filtrata, ad esempio visualizzarla
import matplotlib.pyplot as plt

plt.imshow(filtered_image)
plt.axis('off')
plt.show()
```
